```python
import pandas as pd
import plotly.express as px
```

```python
url = "C:\GitHub\BIA-Team-Bravo\data\clean\Cleaned Employment by industry.csv"
df = pd.read_csv(url)
```

```python
df
```

```python
df.describe()
```

```python
#Values for different industries in Nova Scotia
df[(df['GEO']=='Nova Scotia') & (df['North American Industry Classification System (NAICS)'] != 'North American Industry Classification System (NAICS)')].describe()
```

```python
fig = px.line(
    df[
        (df['North American Industry Classification System (NAICS)'] == 'Goods-producing sector') &
        (df['GEO'] != 'Canada')
    ],
    x='REF_DATE',
    y='VALUE',
    color = 'GEO',
    title = 'Goods-Producing Labour Force',
    labels={
                     "VALUE": "Workforce X 1000",
                     "GEO": "Province",
                     "REF_DATE": "Date"
                 },
)
fig.show()
```

```python
fig = px.line(
    df[
        (df['North American Industry Classification System (NAICS)'] == 'Goods-producing sector') &
        (df['GEO'] == 'Nova Scotia')
    ],
    x='REF_DATE',
    y='VALUE',
    title = 'Nova Scotia Goods-Producing Labour Force',
    labels={
                     "VALUE": "Workforce X 1000",
                     "GEO": "Province",
                     "REF_DATE": "Date"
                 },
)
fig.show()
```

```python
df = df[df.index%2==0]
fig = px.bar(df[(df['GEO'] == 'Nova Scotia') & (df['REF_DATE'] == '2026-03-01')], 
x='North American Industry Classification System (NAICS)', 
y='VALUE',
labels={
                     "VALUE": "Workforce X 1000",
                     'North American Industry Classification System (NAICS)':'Industry'
                 },
title = 'Nova Scotia Workforce March 2026' 
)
fig.show()
```
