```python
import pandas as pd
```

### Reading in CSVs
Creating Dataframes and a dictionary for use in the transformation

```python
url_labor_force = '../data/raw/Labor_Force_Participation_by_State/monthly.csv'
url_construction = '../data/raw/Construction/monthly.csv'
url_education = '../data/raw/Education/monthly.csv'
url_hospitality = '../data/raw/Hospitality/monthly.csv'
url_manufacturing = '../data/raw/Manufacturing/monthly.csv'
```

```python
df_labor_force = pd.read_csv(url_labor_force, index_col=0)
df_construction = pd.read_csv(url_construction, index_col=0)
df_education = pd.read_csv(url_education, index_col=0)
df_hospitality = pd.read_csv(url_hospitality, index_col=0)
df_manufacturing = pd.read_csv(url_manufacturing, index_col=0)
```

```python
dfs = {'us_labor_force':df_labor_force,
'us_construction':df_construction,
'us_education':df_education, 
'us_hospitality':df_hospitality, 
'us_manufacturing':df_manufacturing,}
```

### Indivual Cleaning for Manufacturing
The columns for Manufacturing in District of Columbia and Manufacturing in Alabama are given non-standard ID codes

```python
manufacturing_mapping = {'SMS11000003000000001' : 'DC', 
                         'SMU01000003000000001SA' : 'AL'}
for name, df in dfs.items():
    if name == 'us_manufacturing':
        df.rename(columns=manufacturing_mapping, inplace=True)
        df.drop(columns='SMU11000003000000001SA', inplace=True)
    else:
        pass
```

### Bulk Transformations
Changing each csv's different column code of (State + Survey) into the State Name

```python
for name, df in dfs.items():
    df.rename(columns=lambda x: x[:2], inplace=True)
    df.rename(columns=mapping, inplace=True)
    df.rename_axis(None,inplace=True)
```

### Individual Cleaning
Labor Force has one null State and no information for October 2025

Construction has no information for Tennessee

```python
# df_labor_force cleaning
df_labor_force.drop('Unknown / Not a U.S. state', axis=1, inplace=True)
df_labor_force.drop('2025-10-01', axis=0, inplace=True)
#df_construction cleaning
df_construction.drop('Tennessee', axis=1, inplace=True)
```

### Transposing and Melting
Making it so that rows are states and columns are dates to align with Statistics Canada

```python
dfs = {name: df.T for name, df in dfs.items()}
dfs = {name: df.melt(ignore_index=False) for name, df in dfs.items()}
dfs = {name: df.reset_index(names='State') for name, df in dfs.items()}
dfs = {name: df.rename(columns={'variable':"Date"}) for name, df in dfs.items()}
```

```python
#Final Checking
for name, df in dfs.items():
    print(name)
    display(df)
```

### Saving to CSV

```python
for name, df in dfs.items():
    df.to_csv(f'../data/clean/{name}.csv')
```
