```python
import pandas as pd
```

### Reading in CSVs
Creating Dataframes and a dictionary for use in the transformation

```python
url_GDP_Income = '../Raw Files/US GDP and Personal Income.csv'
url_US_Total_Employment = '../Raw Files/US Total employment By States.csv'
```

```python
US_GDP_Income = pd.read_csv(url_GDP_Income, skiprows=3, skipfooter=6, engine='python')
US_Total_Employment = pd.read_csv(url_US_Total_Employment, index_col=1, skipfooter=4, engine='python')
```

### Cleaning Up and Melting

US GDP and Personal Income

```python
US_GDP_Income = US_GDP_Income.iloc[0:156]
US_GDP_Income['GeoName'] = US_GDP_Income['GeoName'].replace('Alaska *', 'Alaska')
US_GDP_Income.rename(columns={'variable':'Year_Quarter','GeoName':'State'}, inplace=True)
US_GDP_Income = US_GDP_Income.melt(id_vars=['GeoFIPS', 'State', 'LineCode', 'Description'])
US_GDP_Income
```

US Total employment By States

```python
US_Total_Employment.drop(columns = ['2025'], inplace=True)
US_Total_Employment = US_Total_Employment.melt(ignore_index=False)
US_Total_Employment.reset_index(inplace=True)
US_Total_Employment.rename(columns = {'GeoName':'State','variable':'Year','value':'People'}, inplace=True)
```
